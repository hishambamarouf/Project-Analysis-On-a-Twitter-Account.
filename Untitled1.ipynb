## Introduction

"WeRateDogs" is a Twitter account that uploads pictures of different types of dogs and rates them out of 10, the funny thing about it is that it gives ratings that are more than 10, for example 13/10. And for that it has about 9 million followers and it had its fair share of mainstream median coverage.

In this project, we gathered data from this twitter account using python code, along side a prediction algorithm that predicts the dog breed from a picture, cleaned it and analyzed it to find answers to the following questions:
- Which is the most common source used to tweet in the WeRateDogs twitter account.
- What is the most favourited dog breed.
- What are the most common false predictions in the dog prediciton algorithim.
And the following are our findings.



## 1- Which is the most common source used to tweet in the WeRateDogs twitter account

We took the source from over 1800 tweets from the account and found out that most of the dog pictures were tweeted from an Iphone, followed by the Twitter web application then TweetDeck (a social media dashboard application for management of Twitter accounts).

## 2- What is the most favourited dog breed.
